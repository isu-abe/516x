I tried using DBSCAN clustering for the digit data set since it was an unsupervised learning tool where the k did not have to be specified.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.datasets import load_digits
from sklearn.metrics import accuracy_score
from scipy.stats import mode
from sklearn.metrics import confusion_matrix
```

```python
digits = load_digits()
print(digits.data.shape)
```

When my epsilon value is too low, algorithm can't find a good cluster the data and labels everything as noise.

```python
labels=DBSCAN(eps=2,min_samples=8).fit_predict(digits.data)
nClusters=len(set(labels))-(1 if -1 in labels else 0)
print(nClusters)
print(labels.shape)
```

When the minimum samples required to form a core is too low, every data sample becomes a core

```python
labels=DBSCAN(eps=3,min_samples=1).fit_predict(digits.data)
nClusters=len(set(labels))-(1 if -1 in labels else 0)
print(nClusters)
```

Mean of each of the 8x8 pixel value was used to visualize the resulting cluster. Some results such 0 and 7 looks recognizable however 2 and 4 appear to be swapped, 6 and 9 looks like 9 and 8 looks like a warped 1. Best result I could get from DBSCAN was 49% accuracy.

```python
labels=DBSCAN(eps=23.25,min_samples=8).fit_predict(digits.data)
nClusters=len(set(labels))-(1 if -1 in labels else 0)
print(nClusters)

labelsAct = np.zeros_like(labels)
for i in range(nClusters):
    mask = (labels == i)
    labelsAct[mask] = mode(digits.target[mask])[0] #Takes the mode of the true label
print(accuracy_score(digits.target, labelsAct))

fig, ax = plt.subplots(2, 5, figsize=(9, 5))
core=np.unique(labels)
core=core[core>-1]

for axi, i in zip(ax.flat, core):
    mask=(labels==i)
    image=np.mean(digits.images[mask],axis=0)
    axi.set(xticks=[], yticks=[])
    axi.imshow(image, interpolation='nearest', cmap=plt.cm.binary)
    axi.text(0.05, 0.05, str(i), color='green')
```

```python
confuse = confusion_matrix(digits.target, labelsAct)
sns.heatmap(confuse.T, square=True, annot=True, fmt='d',cbar=False, 
            xticklabels=digits.target_names,
            yticklabels=digits.target_names)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

Confusion matrix shows how the model did. There seems to be a lot of false predictions especially wrong predictions of 6.

I found an example where data was dimensionally reduced and scaled. I tried it to see if it improved the model.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca = pca.fit(digits.data)
digitsPCA = pca.transform(digits.data)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler = scaler.fit(digitsPCA)
scaled = scaler.transform(digitsPCA)

labels=DBSCAN(eps=0.125,min_samples=20).fit_predict(scaled)
nClusters=len(set(labels))-(1 if -1 in labels else 0)
print(nClusters)

labelsAct = np.zeros_like(labels)
for i in range(nClusters):
    mask = (labels == i)
    labelsAct[mask] = mode(digits.target[mask])[0] #Takes the mode of the true label
print(accuracy_score(digits.target, labelsAct))

fig, ax = plt.subplots(2, 5, figsize=(9, 5))
core=np.unique(labels)
core=core[core>-1]

for axi, i in zip(ax.flat, core):
    mask=(labels==i)
    image=np.mean(digits.images[mask],axis=0)
    axi.set(xticks=[], yticks=[])
    axi.imshow(image, interpolation='nearest', cmap=plt.cm.binary)
    axi.text(0.05, 0.05, str(i), color='green')
```

```python
confuse = confusion_matrix(digits.target, labelsAct)
sns.heatmap(confuse.T, square=True, annot=True, fmt='d',cbar=False, 
            xticklabels=digits.target_names,
            yticklabels=digits.target_names)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

Scaling it changed the epsilon as distance was scaled to a ratio.

### Discussion
On DBSCAN, minimum sample value increases the rigor required to be considered a core cluster whereas epsilon increases the radius to find mimum samples which increases the chances of a sample becoming a core for a cluster. DBSCAN models seem to predic a few numbers well at a time but there always seem to be a few clusters that looks the same.

As seen in the clustering comparison chart below, DBSCAN is good at seperating clusters based on how closely points stick together, but not on data sets where clusters might overlap a little.


Treating the image pixels as 64 seperate dimensions would result in some overlapping groups which as seen in the 3rd and last row of the chart could lead to overgenerallising in the clusters. In my attempt, it looks like model tend to overgeneralized on zero and the rest of the clusters were whatever small clusters that were left.

I think DBSCAN would be more useful in feature detection within an image. Dataset had clusters of varying density in pixel placement and a high dimensionality both of which DBSCAN is bad at dealing with.

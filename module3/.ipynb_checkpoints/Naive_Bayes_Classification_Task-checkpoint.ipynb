## Task:  classify origin of wine based on physio-chemical analysis data. 

You are provided data that are the results of a chemical analysis of wines grown in the same region in Italy but derived from three different cultivars. The analysis determined the quantities of 13 constituents found in each of the three types of wines.

Details can be [**found here**](http://archive.ics.uci.edu/ml/datasets/Wine).

```python
#Try Flavanoids and Total phenols
#scatter plot
plt.scatter(df['Total phenols'],df['Flavanoids'],c=df['Class'],edgecolors='k')
plt.xlabel('Total phenols')
plt.ylabel('Flavanoids')
plt.title('Flavanoids vs Total phenols')

#Linear regression
linModel = LinearRegression(fit_intercept=True)
x_data = df['Total phenols'][:, np.newaxis]
linModel.fit(x_data, df['Flavanoids'])
yfit = linModel.predict(x_data)

#Fitted line plot
plt.plot(df['Total phenols'],yfit,'r')

#Calculate and print correlation value
r_2=metrics.r2_score(df['Flavanoids'],yfit)
print(f'r^2 value is: {r_2}')
print('Yellow is 3, Green is 2, Purple is 1')
```

####  Naive Bayes Classification

```python
from sklearn.naive_bayes import GaussianNB
```

Use [Naive Bayes Classification](https://scikit-learn.org/stable/modules/generated/sklearn.naive_bayes.GaussianNB.html) to create a model to classify wine base on attributes.  Justify how good the model is for the wine classification.  Note that some of the metrics we've used in class are only for *binary* classifications, so may not eb applicable here.


```python
#Split data into 2 data sets each containing 50% of the data based odd and even index
#1 set for training, 1 set for testing
dfeven=df.iloc[::2].reset_index()
dfodd=df.iloc[1::2].reset_index()
```

```python
#Model 1: Total phenols & Flavanoids training
clf1 = GaussianNB()
clf1.fit(dfeven[['Total phenols','Flavanoids']],dfeven['Class'])

#Prediction test
y1_predict=clf1.predict(dfodd[['Total phenols','Flavanoids']])
print(f"Mean accuracy of model is {metrics.accuracy_score(dfodd['Class'], y1_predict)}")

#Confusion Metric
cm1=metrics.confusion_matrix(dfodd['Class'], y1_predict,labels=[1, 2, 3])
cmdf1 = pd.DataFrame(cm1,index=['1', '2','3'], columns=['1','2','3'])
print("The confusion matrix looks like following...\n")
print(cmdf1)

#Classification Report
print(f"\nCalssification report \n{metrics.classification_report(dfodd['Class'], y1_predict)}")      

#Graph comparison of actual class vs predicted
plt.scatter(dfodd['Total phenols'],dfodd['Flavanoids'],c=dfodd['Class'],edgecolors='k')
plt.xlabel('Total phenols')
plt.ylabel('Flavanoids')
plt.title('Actual')
      
plt.figure()
plt.scatter(dfodd['Total phenols'],dfodd['Flavanoids'],c=y1_predict,edgecolors='k')
plt.xlabel('Total phenols')
plt.ylabel('Flavanoids')
plt.title('Predicted')
```

```python
#Model 2: Color intensity & Proline
clf2 = GaussianNB()
clf2.fit(dfeven[['Color intensity','Proline']],dfeven['Class'])

#Prediction test
y2_predict=clf2.predict(dfodd[['Color intensity','Proline']])
print(f"Mean accuracy is {metrics.accuracy_score(dfodd['Class'], y2_predict)}")
      
#Confusion Metric
cm2=metrics.confusion_matrix(dfodd['Class'], y2_predict,labels=[1, 2, 3])
cmdf2 = pd.DataFrame(cm2,index=['1', '2','3'], columns=['1','2','3'])
print("The confusion matrix looks like following...\n")
print(cmdf2)

#Classification Report
print(f"\nCalssification report \n{metrics.classification_report(dfodd['Class'], y2_predict)}") 
      
#Graph comparison of actual class vs predicted
plt.scatter(dfodd['Color intensity'],dfodd['Proline'],c=dfodd['Class'],edgecolors='k')
plt.xlabel('Color intensity')
plt.ylabel('Proline')
plt.title('Actual')

plt.figure()
plt.scatter(dfodd['Color intensity'],dfodd['Proline'],c=y2_predict,edgecolors='k')
plt.xlabel('Color intensity')
plt.ylabel('Proline')
plt.title('Predicted')
```

```python
#Model 3: Color intensity & Flavanoids
clf3 = GaussianNB()
clf3.fit(dfeven[['Color intensity','Flavanoids']],dfeven['Class'])

#Prediction test
y3_predict=clf3.predict(dfodd[['Color intensity','Flavanoids']])
print(f"Mean accuracy is {metrics.accuracy_score(dfodd['Class'], y3_predict)}")

#Confusion Metric
cm3=metrics.confusion_matrix(dfodd['Class'], y3_predict,labels=[1, 2, 3])
cmdf3 = pd.DataFrame(cm3,index=['1', '2','3'], columns=['1','2','3'])
print("The confusion matrix looks like following...\n")
print(cmdf3)

#Classification Report
print(f"\nCalssification report \n{metrics.classification_report(dfodd['Class'], y3_predict)}") 

#Graph comparison of actual class vs predicted
plt.scatter(dfodd['Color intensity'],dfodd['Flavanoids'],c=dfodd['Class'],edgecolors='k')
plt.xlabel('Color intensity')
plt.ylabel('Flavanoids')
plt.title('Actual')

plt.figure()
plt.scatter(dfodd['Color intensity'],dfodd['Flavanoids'],c=y3_predict,edgecolors='k')
plt.xlabel('Color intensity')
plt.ylabel('Flavanoids')
plt.title('Predicted')
```

```python
#Model 4: Alcohol& Flavanoids
clf4 = GaussianNB()
clf4.fit(dfeven[['Alcohol','Flavanoids']],dfeven['Class'])

#Prediction test
y4_predict=clf4.predict(dfodd[['Alcohol','Flavanoids']])
print(f"Mean accuracy is {metrics.accuracy_score(dfodd['Class'], y4_predict)}")

#Confusion Metric
cm4=metrics.confusion_matrix(dfodd['Class'], y4_predict,labels=[1, 2, 3])
cmdf4 = pd.DataFrame(cm4,index=['1', '2','3'], columns=['1','2','3'])
print("The confusion matrix looks like following...\n")
print(cmdf4)

#Classification Report
print(f"\nCalssification report \n{metrics.classification_report(dfodd['Class'], y4_predict)}") 
      
#Graph comparison of actual class vs predicted
plt.scatter(dfodd['Alcohol'],dfodd['Flavanoids'],c=dfodd['Class'],edgecolors='k')
plt.xlabel('Alcohol')
plt.ylabel('Flavanoids')
plt.title('Actual')

plt.figure()
plt.scatter(dfodd['Alcohol'],dfodd['Flavanoids'],c=y4_predict,edgecolors='k')
plt.xlabel('Alcohol')
plt.ylabel('Flavanoids')
plt.title('Predicted')
```

From metrics, best accuracy came from alcohol, flavanoids and color intensity, flavanoids combination. Accuracy for both these sets are close to 90%. For model 3 recall of class 3 wine is 100%. Model 3 and 4 are good. Model 2 is slightly lower in accuracy compared to Model 3 and 4. Model 1 isn't very good.

Interesting to note that model with highest linear fit had the worst Beyesian model fit.
